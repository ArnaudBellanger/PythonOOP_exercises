```python
class Purse():
    
    def __init__(self):
        #     variable
        self.balance = 10.0
    
#     method
    def  debit (self, amount):
        self.balance -= amount
    
    def credit (self, amount):
        self.balance += amount
    
    def get_balance (self):
        return self.balance
```
